```python
from numpy.random import randn
```

```python
df=pd.DataFrame(dict(time=np.arange(500), value=randn(500).cumsum()))
df.head()
```

```python
sns.relplot(x='time',y='value', kind='line', data=df, sort=True)
```

```python
df=pd.DataFrame(randn(500,2).cumsum(axis=0), columns=['time', 'value'])
df.head()
```

```python
## in this data the values of the data are not sorted. Hence, sort=TRUE/ False will work.

sns.relplot(x='time',y='value', kind='line', data=df, sort=True)

# this fig has sorted
```

```python
sns.relplot(x='time',y='value', kind='line', data=df, sort=False)
#this fig is not sorted
```

```python
dots = sns.load_dataset('dots').query("align=='dots'")
dots.head()
```

```python
sns.relplot(x='time', y='firing_rate', data=dots, kind='line')
```

```python
sns.relplot(x='time', y='firing_rate', data=dots, kind='line', hue='coherence')
```

```python
sns.relplot(x='time', y='firing_rate', data=dots, kind='line', hue='coherence', style='choice')
```

```python
palette=sns.cubehelix_palette(light=0.8, n_colors=6)


sns.relplot(x='time', y='firing_rate', data=dots, kind='line', hue='coherence', style='choice', palette=palette)
```

```python
sns.relplot(x='time', y='firing_rate', data=dots, kind='line', size='coherence', style='choice', palette=palette, sizes=(1, 5))
```
